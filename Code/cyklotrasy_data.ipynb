# Cyklotrasy - dataset

V tomto notebooku sa volá API na získanie dát zaznamenaných snímačmi na cyklotrasách v Bratislave.

```python
import requests as r
import pandas as pd
```

```python
url = 'https://services8.arcgis.com/pRlN1m0su5BYaFAS/arcgis/rest/services/cyklisti2_0/FeatureServer/0/query?where=1%3D1&outFields=*&outSR=4326&f=json'
params = {
    "resultRecordCount": 1000,
}
```

```python
all_records = []
offset = 0

while True:
    params["resultOffset"] = offset

    response = r.get(url, params=params)
    response.raise_for_status()
    data = response.json()

    records = data.get("features", [])
    all_records.extend(records)
    
    if len(all_records) % 10000 == 0:
        print(f"Fetched {len(all_records)} records...")

    # loop končí, keď nie je batch naplnený
    if len(records) < params["resultRecordCount"]:
        break
    
    # offset += počet záznamov v poslednom dotaze
    offset += params["resultRecordCount"]
```

```python
print("Total records:", len(all_records))
```

```python
df = pd.json_normalize(all_records)
```

```python
df.to_csv('../Data/cyklotrasy_data_bratislava_raw.csv', index=False)
```
